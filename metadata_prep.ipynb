```python
import os
import pandas as pd

def create_img_label_csv(image_folder, mask_folder, csv_file):
    # Initialize an empty DataFrame
    df = pd.DataFrame(columns=['image_path', 'label'])
    
    # Check if the CSV file exists
    if os.path.isfile(csv_file):
        # If CSV file exists, read it into DataFrame
        df = pd.read_csv(csv_file)
    
    # Iterate through mask files
    for mask_filename in os.listdir(mask_folder):
        # Extract the image name from the mask filename
        image_name = mask_filename.split("_label")[0]
        
        # Get the full path to the mask file
        mask_path = os.path.join(mask_folder, mask_filename)
        
        # Determine the label based on the mask filename
        label = int(mask_filename.split("_")[-1].split(".")[0][-1])
        
        # Construct the full path to the corresponding image
        image_path = os.path.join(image_folder, image_name) + '.png'
        
        # Append image path and label to DataFrame
        df = df.append({'image_path': image_path, 'label': label}, ignore_index=True)
    
    # Save DataFrame to CSV file
    df.to_csv(csv_file, index=False)

# Example usage:
image_folder = 'artifact/images'
mask_folder = 'artifact/masks'
csv_file = 'metadata.csv'

create_img_label_csv(image_folder, mask_folder, csv_file)
```
